```python
sentencesA=dfA['given_name']
```

```python
sentencesA.dropna(inplace=True)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
import collections
mem=collections.defaultdict()
embededA=[]
for i in sentencesA:
  if i not in mem:
    print(i)
    mem[i]=model.encode(i)
  embededA.append([i,mem[i]])
```

```python
print(embededA[0:5])
```

```python
sentencesB=dfB['given_name']
sentencesB.dropna(inplace=True)
embededB=[]
embededB1=[]
for i in sentencesB:
  if i not in mem:
    print(i)
    mem[i]=model.encode(i)
  embededB.append([i,mem[i]])
  embededB1.append(mem[i])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

```

```python
embededBpd=pd.DataFrame(embededB,columns=['given_name','encoding'])
```

```python
embededBpd.head()
```

```python
out=cosine_similarity([embededA[0][1]],embededB1)
import numpy as np
out=np.squeeze(out, axis=0)
```

```python
embededBpd['sim']=out
```

```python
embededBpd.sort_values(by=['sim'],ascending=False,inplace=True)
```
