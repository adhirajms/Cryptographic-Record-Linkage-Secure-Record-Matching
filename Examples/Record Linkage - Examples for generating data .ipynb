# Generate random binary comparison vectors.

```python
import recordlinkage
import pandas as pd
```

```python
from recordlinkage.datasets import binary_vectors
```

```python
features, links = binary_vectors(n = 10, n_match = 3, return_links=True)
print("Features")
display(features.sort_index().head(10))
print("Links")
display(links)
```

```python
features['tot'] = features['c_1'] + features['c_2'] + features['c_3'] + features['c_4'] + features['c_5'] + \
                     features['c_6'] + features['c_7'] + features['c_8']

features.sort_values(by='tot',ascending=False)
```
